```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/fake_or_real_news.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
label={'FAKE':0,'REAL':1}

df.label=[label[i] for i in df.label]
print(df)
```

```python
x=df.drop(['label'],axis=1)
y=df['label']
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Embedding
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
voc_size=7000
```

```python
messages=x.copy()
```

```python
messages.reset_index(inplace=True)
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('stopwords')
  
```

```python
corpus=[]

for i in range(len(messages)):
  print(i)

  set=re.sub('[^a-zA-Z]',' ',messages['title'][i])
  set=re.sub('#\w+','',messages['title'][i])
  set=set.lower()
  set=set.split()
  set=[w for w in set if not w in stopwords.words('english')]
  set=' '.join(set)
  corpus.append(set)
```

```python
corpus
```

```python
one_hot=[one_hot(words,voc_size)for words in corpus]
one_hot
```

```python
sent_len=20

embeddings=pad_sequences(one_hot,padding='post',maxlen=sent_len)
print(embeddings)
```

```python
len(embeddings)
```

```python
embeddings[0]
```

```python
embedding_vector=40

from tensorflow.keras.models import Sequential

model=Sequential([
                  Embedding(voc_size,embedding_vector,input_length=sent_len),
                  LSTM(100),
                  tf.keras.layers.Dense(1,activation='sigmoid')
])

model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

model.summary()
```

```python
len(embeddings),y.shape
```

```python
import numpy as np
x_final=np.array(embeddings)
y_final=np.array(y)
```

```python
x_final.shape,y_final.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x_final,y_final,test_size=0.3,random_state=20)
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```
